```python
from ultralytics import YOLO
```

```python
# Load YOLOv8 nano model (lightweight for CPU)
model = YOLO("yolov8n.pt")  # auto-downloads if not present
```

```python
# Use a short sample video or webcam
video_path = "traffic_sample.mp4"  # replace with your video path or 0 for webcam
cap = cv2.VideoCapture(video_path)
```

```python
frame_count = 0
max_frames = 500  # Process about ~20 seconds of video

while cap.isOpened() and frame_count < max_frames:
    ret, frame = cap.read()
    if not ret:
        break

    # Inference
    results = model(frame)

    # Annotated frame
    annotated = results[0].plot()

    # Show frame (in a separate window)
    cv2.imshow("YOLOv8 Vehicle Detection", annotated)

    # Press 'q' to stop early
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

    frame_count += 1

# Release resources
cap.release()
cv2.destroyAllWindows()
print("✅ Video processing completed safely!")
```

# SECTION 2: TRAFFIC LIGHT CONTROL WITH RL

```python
import numpy as np

import os 
import gym 
from stable_baselines3 import PPO
from stable_baselines3.common.vec_env import DummyVecEnv
from stable_baselines3.common.evaluation import evaluate_policy
```

```python
# Simple traffic environment (simulate cars with random flow)
class SimpleTrafficEnv(gym.Env):
    def __init__(self):
        super(SimpleTrafficEnv, self).__init__()
        self.action_space = gym.spaces.Discrete(2)  # 0 = keep, 1 = switch light
        self.observation_space = gym.spaces.Box(low=0, high=100, shape=(2,), dtype=np.float32)
        self.reset()

    def reset(self):
        self.queue = [np.random.randint(0, 50), np.random.randint(0, 50)]
        return np.array(self.queue, dtype=np.float32)

    def step(self, action):
        # Reduce cars in green lane, add random new cars in red lane
        if action == 0:
            self.queue[0] = max(0, self.queue[0] - np.random.randint(5, 15))
            self.queue[1] += np.random.randint(0, 10)
        else:
            self.queue[1] = max(0, self.queue[1] - np.random.randint(5, 15))
            self.queue[0] += np.random.randint(0, 10)
        reward = -sum(self.queue)
        done = False
        return np.array(self.queue, dtype=np.float32), reward, done, {}
```

```python
env = DummyVecEnv([lambda: SimpleTrafficEnv()])
model = PPO("MlpPolicy", env, verbose=0)
```

```python
model.learn(total_timesteps=5000)
```

```python
print("✅ RL training complete. Try predicting:")
print(model.predict(env.reset()))
```

# SECTION 3: TRAFFIC FORECASTING WITH GNN

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
```

```python
# Create synthetic graph traffic data
num_nodes = 5
time_steps = 50
features = torch.rand((time_steps, num_nodes, 1))
targets = torch.roll(features, shifts=-1, dims=0)
```

```python
class GNN(nn.Module):
    def __init__(self, in_feats, hidden, out_feats):
        super(GNN, self).__init__()
        self.fc1 = nn.Linear(in_feats, hidden)
        self.fc2 = nn.Linear(hidden, out_feats)
    def forward(self, x):
        x = F.relu(self.fc1(x))
        return self.fc2(x)
```

```python
model_gnn = GNN(1, 16, 1)
optimizer = optim.Adam(model_gnn.parameters(), lr=0.01)
loss_fn = nn.MSELoss()
```

```python
for epoch in range(100):
    optimizer.zero_grad()
    out = model_gnn(features)
    loss = loss_fn(out, targets)
    loss.backward()
    optimizer.step()
    print(f"Epoch {epoch+1}, Loss: {loss.item():.4f}")

print("✅ GNN training complete.")
```
